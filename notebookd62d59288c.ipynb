# Analyzing for iris petals data


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris_df = pd.read_csv("../input/seaborn-data-iris/seaborn_data_iris.csv")
iris_df.head(10)
```

### find the species that exist in the data 

```python
iris_df["species"].value_counts()
```

### There are 3 species in the dataset

Let's plot each of them with thier sepal's & petal's lengtha & width and compare between the species.

```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "sepal_width")
```

### Got From Graph:
- sepal width in the 3 species range between 2.25 - 4.25 mm, but they vary in distribution:

    - satosa's sepal width distribution is skewed to the right, which means more sepals are wider, and has the widest sepals.

    - versicolor's sepal width distribution is skewed to the left, which means more sepals are wider, and has the narrowest sepals.

    - virginica's sepal width distribution looks like a centered normal distribution, and the middle value is almost 3.1 mm


```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "sepal_length")
```

### Got From Graph:
- satosa's sepal length distribution is skewed to the left, and has the shortest petals.
- versicolor's sepal length distribution does not have a specefic shape.
- virginica's sepal length distribution ranges between 5 - 8 mm length, and has the longest sepals.


```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "petal_width")
```

### Got From Graph:
- satosa's petal's width distribution is skewed to the left, and has very narrow petals, between 0.1 - 0.6 mm
- versicolor's petals width distribution does not have a specefic shape.
- virginica's sepal width distribution is almost equaly distributed between the values 1.5 - 2.5.


```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "petal_length")
```

### Got From Graph:
- satosa's petal's length distribution is skewed to the left, and has very short petals, between 0.5 - 2 mm
- versicolor's petals width distribution is skewed to the right between the values: 3 - 5 mm
- virginica's petals width distribution is skewed to the left between the values: 2.5 - 6.5 mm



---

After comparing each dimention of the sepal and petal between the different species, now it is the time to compare between the ration of width to length of the sepals and petals, and try to get an idea about the shapes of them.

```python
iris_df["s_width to length ratio"] = iris_df["sepal_width"]/iris_df["sepal_length"]
iris_df["p_width to length ratio"] = iris_df["petal_width"]/iris_df["petal_length"]
```

```python
iris_df.head()
```

```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "s_width to length ratio")
```

### This data gives us insights about the general shap of these species sepals

- the setosa's sepals seem to have the most "circular" shape, as the ratio is the closest to 1 which agrees with our previous notes, because its' sepal's are wider and shorter in general.
- the other 2 species sepals seem to be similar in shape, that ranges between a very elongated shape to a fair one.

```python
g = sns.FacetGrid(iris_df, col="species")
g.map(sns.histplot, "p_width to length ratio")
```

### as for petals, the data is not as friendly
- setosa's petals are generally the narrowest.
- vesicolor's petals ratio is usually very close to or equal to 3, which means that there is not much variaty in their shape, they are very similar.
- virginica's petals are the least narrow among all of them.
